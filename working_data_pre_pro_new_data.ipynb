```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0,255))
```

```python
RGB_data = scaler.fit_transform(df2[['Red','Green','Blue']])
RGB_data
```

```python
rgb_df = pd.DataFrame(RGB_data,columns=["Red","Green","Blue"])
rgb_df = rgb_df.astype({"Red":'int',"Green":'int',"Blue":'int'})
rgb_df['X'] = df2['X']
rgb_df['Y'] = df2['Y']
rgb_df['Z'] = df2['Z']
rgb_df['Classification'] =df2['Classification']
rgb_df['Intensity'] = df2['Intensity']
rgb_df = rgb_df.astype({'Classification':'int'})
rgb_df
```
